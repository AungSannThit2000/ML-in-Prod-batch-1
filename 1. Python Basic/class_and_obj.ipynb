```python
from model import Item, Categories, SaleItem





cat1 = Categories('03-06-2024','cat_001','Local coffee')

i1 = Item ('12-06-2024','i_001','super_coffee',"0000",3900,3500,cat1)
i2 = Item ('17-06-2024','i_002','sunday_coffee',"0000",3450,3000,cat1)



```

```python
from model import SaleItem


sale1=  SaleItem(customer_id='CU_000001',created_date='15-06-2024',item=i1 )
sale2=  SaleItem(customer_id='CU_000001',created_date='21-06-2024',item=i1 )
sale3=  SaleItem(customer_id='CU_000001',created_date='21-07-2024',item=i2 )
total_sale = [sale1,sale2,sale3]
```

```python
category,date_list ,profit_list = [],[],[]
for cur_obj in total_sale:
    category_name, profit = cur_obj.getCategoryAndProfit()
    category.append(category_name),
    profit_list.append(profit)
    date_list.append(cur_obj.created_date)

print(category)
print(profit_list)
print(date_list)
```

```python
import matplotlib.pyplot as plt

plt.plot(date_list, profit_list, label = "line 1", linestyle="-.") 
#plt.plot(profit_list, , label = "line 2", linestyle="--") 
plt.show()
```
